## Analyzing profits and sales acquired by each ship mode

## Visualizing the profits generated by each ship mode

```python
ship_mode=data.groupby(data['Ship Mode']).Profit.sum().plot.bar(xlabel='Ship Mode', ylabel='Total Profits Generated',title='Profits generated by Ship Mode')
```

### We can clearly see that the Standard Class has acquired the maximum profits in comparison to other ship modes and the Same Day ship mode has generated the lowest profits

## Visualizing the total sales generated by ship modes

```python
ship_mode_sales=data.groupby(data['Ship Mode']).Sales.sum().plot.bar(title='Total Sales genereated by each ship mode',ylabel='Sales',color='g',figsize=(9,6))
```

### We can see that the First Class and Second Class ship modes have nearly equal amount of sales to each other and Standard class is yet again highest in sales. 

```python
sales = data.groupby('Ship Mode').Sales.sum()
profit = data.groupby('Ship Mode').Profit.sum()
sales.plot.bar()
profit.plot.bar(color='r',figsize=(8,9))
plt.xlabel('Ship Mode')
plt.ylabel('Sales & Profits')
plt.title('Sales and Profits generated by each Ship Mode')
plt.xticks(rotation=0)
plt.show()
```

### This graph shows both sales and profits w.r.t. Ship Mode.It can be clearly seen that the customers prefer the Standard class more and hence, it generates profits accordingly.
